```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 3: Select Features for Clustering
features = [
    'Age',
    'Best Option: Mutual Funds',
    'Best Option: Equity Market',
    'Best Optioin: Debentures',
    'Best Option: Government Bonds',
    'Best Option: Fixed Deposits',
    'Best Option: Public Provident Fund',
    'Best Option: Gold'
]
df_cluster = df[features].copy()
```

```python
# Step 4: Run KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(df_cluster)
```

```python
# Step 5: PCA for 2D Visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_cluster[features])
df_cluster['PCA1'] = pca_components[:, 0]
df_cluster['PCA2'] = pca_components[:, 1]
```

```python
# Step 6: Visualize the Clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cluster, x='PCA1', y='PCA2', hue='Cluster', palette='Set2')
plt.title("Investor Segmentation using K-Means Clustering")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
```

```python
# Sample function to map clusters to persona
def label_persona(cluster):
    if cluster == 0:
        return "Cautious Young Investor"
    elif cluster == 1:
        return "Aggressive Growth Seeker"
    elif cluster == 2:
        return "Experienced Risk-Balancer"
    else:
        return "Unknown"

# Streamlit App Title
st.title("Investor Persona Predictor")

st.markdown("""
Enter your investment profile below to discover your investor persona:
""")

# Input fields
age = st.slider("Age", 18, 65, 30)
mutual_fund = st.slider("Rank Mutual Funds (1-7)", 1, 7, 4)
equity_market = st.slider("Rank Equity Market (1-7)", 1, 7, 4)
debentures = st.slider("Rank Debentures (1-7)", 1, 7, 4)
gov_bonds = st.slider("Rank Government Bonds (1-7)", 1, 7, 4)
fixed_deposits = st.slider("Rank Fixed Deposits (1-7)", 1, 7, 4)
ppf = st.slider("Rank Public Provident Fund (1-7)", 1, 7, 4)
gold = st.slider("Rank Gold (1-7)", 1, 7, 4)

# Load original dataset for training clustering model
@st.cache_data
def load_data():
    df = pd.read_csv("Final Data.csv")
    features = [
        'Age',
        'Best Option: Mutual Funds',
        'Best Option: Equity Market',
        'Best Optioin: Debentures',
        'Best Option: Government Bonds',
        'Best Option: Fixed Deposits',
        'Best Option: (Public Provident Fund]',
        'Best Option: Gold'
    ]
    return df[features]

df = load_data()

# Train the clustering model
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df_scaled)

# Predict user input
user_data = np.array([[age, mutual_fund, equity_market, debentures, gov_bonds, fixed_deposits, ppf, gold]])
user_scaled = scaler.transform(user_data)
cluster = kmeans.predict(user_scaled)[0]
persona = label_persona(cluster)

# Show result
st.subheader("Predicted Investor Persona:")
st.success(f"{persona}")
```
